```python
# Datapakker importeres.

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import requests
import seaborn as sns
```

## Datasæt: antal uddannede pædagoger

```python
# Datasæt indlæses.

antal_uddannede1 = pd.read_excel ('Antal uddannede pædagoger1.xlsx', sep = ',')
antal_uddannede1

# Vi ændrer dataframe så kolonnen med uddannelsessteder i København er index.
antal_uddannede = antal_uddannede1.set_index('Unnamed: 0')
antal_uddannede

antal_uddannede.index.names = ['Uddannelsessteder i Kbh']

antal_uddannede
```

## Datasæt: antal børn i KK

```python
# Importere yderligere data.
antal_boern1 = pd.read_excel ('Antal børn i KK.xlsx', sep = ',')

# Vi ændrer dataframe så kolonnen med børnenes alder er index.
antal_boern = antal_boern1.set_index('Unnamed: 0')

antal_boern.index.names = ['Alder']
antal_boern.columns.names = ['Årstal']

antal_boern
```

## Datasæt: befolkningsfremskrivning for børn i KK

```python
# importere yderligere data.
befolkningsfremskrivning1 = pd.read_excel ('Befolkningsfremskrivning KK 0-6 år.xlsx', sep = ',')

# Vi ændrer dataframe så kolonnen med børnenes alder er index.
befolkningsfremskrivning = befolkningsfremskrivning1.set_index('Unnamed: 0')

befolkningsfremskrivning.index.names = ['Alder']
befolkningsfremskrivning.columns.names = ['Årstal']

befolkningsfremskrivning
```


## Sammentæller kolonner i datasæt: antal_boern og antal_uddannede tabeller 

```python
# tager summen af alle kolonner og navngiver i alt
sums_boern = antal_boern.sum().rename('Børn i alt')

# indsætter i alt-rækken i antal_boern dataframe
boern_sum = antal_boern.append(sums_boern)
boern_sum
```

```python
# tager summen af alle kolonner i antal_uddannede og navngiver 'i alt'
sums_uddannede = antal_uddannede.sum().rename('Uddannede i alt')

# indsætter i alt-rækken i antal_uddannede dataframe
udd_sum = antal_uddannede.append(sums_uddannede)
udd_sum
```

## Merge af antal_børn og antal_uddannede i alt
Benytter 'i alt' rækkerne og de årstal, som de to tabeller har til fælles

```python
# concat bruges til at sammenflette boern_sum og udd_sum ud fra de kolonner de to tabeller deler
concat_boern_udd = pd.concat([boern_sum, udd_sum], join='inner', axis=0, sort=False)
concat_boern_udd

# Bytter om på rækker og kolonner for at få en pænere tabel
concat_boern_udd2 = concat_boern_udd.T
concat_boern_udd2
```

### Sammenflettet tabel med sumværdierne fra antal børn og antal uddannede pædagoger

```python
# Laver en tabel med årstal og summen af børn for hver årstal og uddannede pædagoger for hvert årstal (2015-2020)
ialt_boern_udd = concat_boern_udd2[['Børn i alt', 'Uddannede i alt']]

# Reset af index, så Årstal er kolonnenavn
ialt_boern_udd1 = ialt_boern_udd.reset_index()
ialt_boern_udd1
```

# Plot for antal børn og antal pædagoger i KK 2015-2020

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.scatter(x=ialt_boern_udd1['Årstal'], y=ialt_boern_udd1['Børn i alt'])
ax.set(xlabel='Årstal', ylabel='Antal børn 0-6 år i KK', title='Stigning af børn 2015-2020')
plt.ylim((50000,53000))

fig, ax = plt.subplots(figsize=(10, 5))
ax.scatter(x=ialt_boern_udd1['Årstal'], y=ialt_boern_udd1['Uddannede i alt'])
ax.set(xlabel='Årstal', ylabel='Antal uddannede pædagoger i KK', title='Fald i uddannede pædagoger 2015-2020')
```

## Befolkningfremskrivning børn 0-6 år i KK - pæn og ordnet tabel

```python
# tager summen af alle kolonner og navngiver i alt
sums_befolkningsfremskrivning = befolkningsfremskrivning.sum().rename('Børn i alt')

# indsætter i alt-rækken i befolkningsfremskrivning dataframe
befolkningsfremskrivning_sum = befolkningsfremskrivning.append(sums_befolkningsfremskrivning)
befolkningsfremskrivning_sum 

# Bytter rundt på rækker og kolonner
befolkningsfremskrivning_T = befolkningsfremskrivning_sum.T

# Reset af index, så Årstal er kolonnenavn
befolkningsfremskrivning_plot = befolkningsfremskrivning_T.reset_index()
befolkningsfremskrivning_plot
```

## Plot over befolkningsfremskrivning i KK 0-6 år

```python
fig, ax = plt.subplots(figsize=(20, 5))
ax.scatter(x=befolkningsfremskrivning_plot['Årstal'], y=befolkningsfremskrivning_plot['Børn i alt'])
ax.set(xlabel='Årstal', ylabel='Total antal børn 0-6 år i KK', title='Befolkningfremskrivning af børn i KK 2020-2045')
```

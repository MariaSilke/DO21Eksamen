
## Sammentæller kolonner i datasæt: antal_boern og antal_uddannede tabeller 

```python
# tager summen af alle kolonner og navngiver i alt
sums_boern = antal_boern.sum().rename('Børn i alt')

# indsætter i alt-rækken i antal_boern dataframe
boern_sum = antal_boern.append(sums_boern)
boern_sum
```

```python
# tager summen af alle kolonner i antal_uddannede og navngiver 'i alt'
sums_uddannede = antal_uddannede.sum().rename('Uddannede i alt')

# indsætter i alt-rækken i antal_uddannede dataframe
udd_sum = antal_uddannede.append(sums_uddannede)
udd_sum
```

## Merge af antal_børn og antal_uddannede i alt
Benytter 'i alt' rækkerne og de årstal, som de to tabeller har til fælles

```python
# concat bruges til at sammenflette boern_sum og udd_sum ud fra de kolonner de to tabeller deler
concat_boern_udd = pd.concat([boern_sum, udd_sum], join='inner', axis=0, sort=False)
concat_boern_udd

# Bytter om på rækker og kolonner for at få en pænere tabel
concat_boern_udd2 = concat_boern_udd.T
concat_boern_udd2
```

### Sammenflettet tabel med sumværdierne fra antal børn og antal uddannede pædagoger

```python
# Laver en tabel med årstal og summen af børn for hver årstal og uddannede pædagoger for hvert årstal (2015-2020)
ialt_boern_udd = concat_boern_udd2[['Børn i alt', 'Uddannede i alt']]

# Reset af index, så Årstal er kolonnenavn
ialt_boern_udd1 = ialt_boern_udd.reset_index()
ialt_boern_udd1
```

# Plot for antal børn og antal pædagoger i KK 2015-2020

```python
#Plot for udviklingen i antal børn i KK
fig, ax = plt.subplots(figsize=(10, 5))
x1 = ialt_boern_udd1['index']
y1 = ialt_boern_udd1['Børn i alt']
ax.scatter(x1, y1)
ax.plot(x1, y1)
ax.set(xlabel='Årstal', ylabel='Antal børn 0-6 år i KK', title='Udvikling i antal børn 2015-2020 i KK')
plt.ylim((50000,53000))

# Plot for udviklingen i antal uddannede pædagoger i KK
fig, ax = plt.subplots(figsize=(10, 5))
x2 = ialt_boern_udd1['index']
y2 = ialt_boern_udd1['Uddannede i alt']
ax.plot(x2, y2)
ax.scatter(x2, y2)
ax.set(xlabel='Årstal', ylabel='Antal uddannede pædagoger i KK', title='Udvikling i antal uddannede pædagoger 2015-2020 i KK')
```

# Funktion der foreskriver udbud indtil 2030
- I 2020 er udbuddet af pædagoger, pædagogmedhjælpere og -assistenter og dagplejere på 7352
- Tallet er fundet på Danmarks statistik 25/11-21

```python
# Udregning af udbuddet af pædagoger fra 2021 til 2030
# Vi ved allerede at udbuddet af pædagoger i 2020 var på 7352
# Indsættes i dataframe

aar = 2020
udbud20 = 7352

dataUdbud = []

for uddannedeP in range(10):
    uddannedeP = udbud20 * 1.026
    udbud20 = uddannedeP
    aar += 1
    dataUdbud.append({'Årstal': aar, 'Udbud': udbud20})
    dfUdbud1 = pd.DataFrame(dataUdbud)
    

dataUdbud20 = []

dataUdbud20.insert(0, {'Årstal': '2020', 'Udbud': 7352})

dfUdbud = pd.concat([pd.DataFrame(dataUdbud20), dfUdbud1], ignore_index=True)

dfUdbud['Udbud'] = dfUdbud['Udbud'].astype(int)
dfUdbud
```

```python
# Udregning af efterspørgsel af pædagoger fra 2020 til 2030
# Befolkningfremskrivningstabellen over børn i 0-5 års aldereren i KK giver os tallene

# Vi finder efterspørgslen i 2020 vha. minimunsnummeringen på 1:6


b = np.array([52390, 53125, 53881, 54682, 55411, 56325, 57625, 58918, 60353, 61689, 62866])
aar2 = np.array([2020, 2021, 2022, 2023, 2024, 2025, 2026, 2027, 2028, 2029, 2030])

eftersp = b/6

# Laver dataframe over efterspørgsel 
dfEftersp = pd.DataFrame()

dfEftersp['Årstal'] = aar2
dfEftersp['Efterspørgsel'] = eftersp

dfEftersp['Efterspørgsel'] = dfEftersp['Efterspørgsel'].astype(int)
dfEftersp
```

```python
# Udregning af forskellen mellem udbud og efterspørgsel på pædagoger

aar3 = ([2020, 2021, 2022, 2023, 2024, 2025, 2026, 2027, 2028, 2029, 2030])
forskel = dfEftersp['Efterspørgsel'] - dfUdbud['Udbud']

dfForskel = pd.DataFrame()
dfForskel['Årstal'] = aar3
dfForskel['Forskel'] = forskel

dfForskel
```

```python
# merge bruges til at sammenflette fdUdbud og dfEftersp ud fra de kolonner de to tabeller deler

merge_UEF = pd.merge(pd.merge(dfUdbud, dfEftersp, on='Årstal'), dfForskel, on='Årstal')

merge_UEF
```

```python
# Definerer forskellige værdier
x = merge_UEF['Årstal']
y = merge_UEF['Udbud']
z = merge_UEF['Efterspørgsel']
a = merge_UEF['Forskel']

# Plot over udbud og efterspørgsel
fig, ax = plt.subplots(figsize=(10, 5))
plt.scatter(x, y, color='r')
plt.scatter(x, z, color='b')
lineU = plt.plot(x, y, color = 'r', label="Udbud")
lineE = plt.plot(x, z, color = 'b', label="Efterspørgsel")
plt.xlabel('Årstal'), plt.ylabel('Pædagoger')
plt.title('Udvikling af udbud og efterspørgsel af pædagoger frem til 2030')
plt.legend()

# Plot over forskellen i udbud og efterspørgsel
fig, ax = plt.subplots(figsize=(10, 5))
plt.scatter(x, a, color='g')
lineF = plt.plot(x, a, color = 'g', label="Forskel i udbud og efterspørgsel")
plt.xlabel('Årstal'), plt.ylabel('Pædagoger')
plt.title('Udvikling af forskellen mellem udbud og efterspørgsel af pædagoger frem til 2030')

plt.legend()
#plt.show()
```

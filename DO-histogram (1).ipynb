```python
#laver string-procenter om til integers, så vi kan bruge dem
df1['paedpct'] = df1['paedpct'].str.rstrip('%').astype('int')
df1
```

```python
df1['paedpct']
```

```python
ppct = df1['paedpct']
```

```python
fig, ax = plt.subplots(figsize = (20, 10))
ax.set_title('Institutioners andel af pædagogisk personale som er faglærte', fontsize = 25)
ax = sns.distplot(ppct,hist=True)
plt.xlabel('Procentdel af inst. som er udd. pædagoger', fontsize=20)
plt.ylabel('Density', fontsize=20)
sns.despine()
plt.axvline(x=66, ymin=0, ymax=1, color = 'red', label = 'målsætning')
plt.text(67.1,0.04,'målsætning',rotation=0)
```

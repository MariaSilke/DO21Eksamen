
## Finder summen af kolonner i datasæt: antal_boern og antal_uddannede tabeller 

```python
# Finder summen af alle kolonner i datasættet antal_boern og navngiver i alt
sums_boern = antal_boern.sum().rename('Børn i alt')

# Indsætter i 'alt'-rækken i antal_boern dataframe
boern_sum = antal_boern.append(sums_boern)

# Vis datasættet
boern_sum
```

```python
# Tager summen af alle kolonner i datasættet antal_uddannede og navngiver 'i alt'
sums_uddannede = antal_uddannede.sum().rename('Uddannede i alt')

# Indsætter 'i alt'-rækken i antal_uddannede dataframe
udd_sum = antal_uddannede.append(sums_uddannede)

# Vis datasættet
udd_sum
```

## Concat af antal_børn og antal_uddannede i alt
Benytter 'i alt' rækkerne og de årstal, som de to tabeller har til fælles til at kombinere dem i én tabel

```python
# Concat bruges til at sammenflette boern_sum og udd_sum ud fra de kolonner de to tabeller deler
concat_boern_udd = pd.concat([boern_sum, udd_sum], join='inner', axis=0, sort=False)

# Bytter om på rækker og kolonner for at få en pænere tabel
concat_boern_udd2 = concat_boern_udd.T

# Vis datasættet
concat_boern_udd2
```

### Sammenflettet tabel kun med sumværdierne fra antal børn og antal uddannede pædagoger

```python
# Laver en tabel med årstal og summen af børn for hver årstal og uddannede pædagoger for hvert årstal (2015-2020)
ialt_boern_udd = concat_boern_udd2[['Børn i alt', 'Uddannede i alt']]

# Reset af index, så Årstal er kolonnenavn
ialt_boern_udd1 = ialt_boern_udd.reset_index()
ialt_boern_udd1.rename(columns={'index': 'Årstal'}, inplace=True)

# Vis datasættet
ialt_boern_udd1
```

# For-loop der foreskriver udbud indtil 2030
- I 2020 er udbuddet af pædagoger, pædagogmedhjælpere og -assistenter og dagplejere på 7352 - Tallet er fundet på Danmarks statistik 25/11-21
- Man forventer at udbuddet af pædagoger vil stige med 2,6% om året

```python
# Udregning af udbuddet af pædagoger fra 2021 til 2030

# Definerer startværdier
aar = 2020
udbud20 = 7352

dataUdbud = []

# Laver for-loop der tager udgangspunkt i 7352 pædagogisk personale i år 2020 og for hvert år lægger 2,6% oveni
# Data fra loop indsættes i dataFrame
for uddannedeP in range(10):
    uddannedeP = udbud20 * 1.026
    udbud20 = uddannedeP
    aar += 1
    dataUdbud.append({'Årstal': aar, 'Udbud': udbud20})
    dfUdbud1 = pd.DataFrame(dataUdbud)
    
# Inkluderer data fra 2020
dataUdbud20 = []
dataUdbud20.insert(0, {'Årstal': '2020', 'Udbud': 7352})

# Sætter data sammen fra loop og fra 2020 i en dataFrame
dfUdbud = pd.concat([pd.DataFrame(dataUdbud20), dfUdbud1], ignore_index=True)

# Viser tal i hele tal
dfUdbud['Udbud'] = dfUdbud['Udbud'].astype(int)

# Vis datasættet
dfUdbud
```

## Finder efterspørgsel af pædagoger fra 2020-2030

```python
# Befolkningfremskrivningstabellen over børn i 0-5 års aldereren i Københavns Kommune giver os tallene
# Vi finder efterspørgslen i 2020 vha. disse tal og minimunsnummeringen på 1:6


b = np.array([52390, 53125, 53881, 54682, 55411, 56325, 57625, 58918, 60353, 61689, 62866])
aar2 = np.array([2020, 2021, 2022, 2023, 2024, 2025, 2026, 2027, 2028, 2029, 2030])

# Dividerer med 6 så det passer med minimumsnormeringerne
eftersp = b/6

# Laver dataframe over efterspørgsel 
dfEftersp = pd.DataFrame()

dfEftersp['Årstal'] = aar2
dfEftersp['Efterspørgsel'] = eftersp

dfEftersp['Efterspørgsel'] = dfEftersp['Efterspørgsel'].astype(int)

# Vis datasættet
dfEftersp
```

## Hvor meget pædagigisk personale mangler hvert år

```python
# Udregning af forskellen mellem udbud og efterspørgsel på pædagogisk personale

aar3 = ([2020, 2021, 2022, 2023, 2024, 2025, 2026, 2027, 2028, 2029, 2030])
forskel = dfEftersp['Efterspørgsel'] - dfUdbud['Udbud']

dfForskel = pd.DataFrame()
dfForskel['Årstal'] = aar3
dfForskel['Forskel'] = forskel

# Vis datasættet
dfForskel
```

```python
# merge bruges til at sammenflette fdUdbud, dfEftersp og dfForskel ud fra de kolonner de to tabeller deler

merge_UEF = pd.merge(pd.merge(dfUdbud, dfEftersp, on='Årstal'), dfForskel, on='Årstal')

# Vis datasættet
merge_UEF
```

## Plot over udbud, efterspørgsel og forskel

```python
# Definerer forskellige værdier
x = merge_UEF['Årstal']
y = merge_UEF['Udbud']
z = merge_UEF['Efterspørgsel']
a = merge_UEF['Forskel']

# Plot over udbud og efterspørgsel
fig, ax = plt.subplots(figsize=(10, 5))
plt.scatter(x, y, color='r')
plt.scatter(x, z, color='b')
lineU = plt.plot(x, y, color = 'r', label="Udbud")
lineE = plt.plot(x, z, color = 'b', label="Efterspørgsel")
plt.xlabel('Årstal'), plt.ylabel('Pædagogisk personale')
plt.title('Udvikling af udbud og efterspørgsel af pædagogisk personale frem til 2030')
plt.legend()

# Plot over forskellen i udbud og efterspørgsel
fig, ax = plt.subplots(figsize=(10, 5))
plt.scatter(x, a, color='g')
lineF = plt.plot(x, a, color = 'g', label="Forskel i udbud og efterspørgsel")
plt.xlabel('Årstal'), plt.ylabel('Pædagogisk personale')
plt.title('Udvikling af forskellen mellem udbud og efterspørgsel af pædagogisk personale frem til 2030')

plt.legend()
#plt.show()
```

# Pris for ønsket antal pædagogisk personale i 2022

```python
# Finder forskellen mellem udbud og efterspørgsel af pædagogisk personale i 2022.
dfForskel['Forskel'][2]

# Regner 66% ud af 1241 manglende pædagogisk personale
stor_forskel = dfForskel['Forskel'][2]*0.66
stor_forskel

# Regner 34% ud af 1242 manglende pædagogisk personale
lille_forskel = dfForskel['Forskel'][2]-stor_forskel
lille_forskel

# Finder den samlede løn pr. år for en uddannet pædagog
loen_uddannet = loen['Pædagoger']['Samlet løn']*12
loen_uddannet

# Finder den samlede løn pr. år for ufaglært pædagog
loen_ikke_uddannet = loen['Ikke udd personale']['Samlet løn']*12
loen_ikke_uddannet

# Finder prisen for 66% uddannede pædagoger
pris_66_uddannede = int(stor_forskel*loen_uddannet)
pris_66_uddannede

#Finder prisen fra 34% ikke uddannede pædagoger.
pris_34_uddannede = int(lille_forskel*loen_ikke_uddannet)
pris_34_uddannede

i_alt = pris_66_uddannede+pris_34_uddannede

# Vis omkostning
i_alt

# Det vil koste KK 528.719.675,- om året, hvis 66% skal være uddannet pædagoger og 34% skal være ikke uddannet.
```

```python
# Tabel over de egentlige omkostninger ved at leve op til de to mål 
# sammenlignet med de penge der er afsat til det.

forskel3 = i_alt - 450000000

data1 = {'Årstal':[2022], 'Omkostninger': [i_alt], 'Rådighedsbeløb':[450000000], 'Mangler i kr.': [forskel3]}
omkost2 = pd.DataFrame(data = data1)

# Viser datasættet
omkost2
```
